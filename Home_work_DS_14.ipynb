```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
num_words = 10000  
maxlen = 200  

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_words)
```

```python
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
```

```python
model_rnn = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=num_words, output_dim=128, input_length=maxlen),
    tf.keras.layers.SimpleRNN(128, return_sequences=False),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_rnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_rnn.summary()
```

```python
model_lstm = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=num_words, output_dim=128, input_length=maxlen),
    tf.keras.layers.LSTM(128, return_sequences=False),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_lstm.summary()
```

```python
model_bi_lstm = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=num_words, output_dim=128, input_length=maxlen),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_bi_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_bi_lstm.summary()
```

```python
model_deep_lstm = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=num_words, output_dim=128, input_length=maxlen),
    tf.keras.layers.LSTM(128, return_sequences=True),
    tf.keras.layers.LSTM(128),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_deep_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_deep_lstm.summary()
```

```python
batch_size = 64
epochs = 5

history_rnn = model_rnn.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=epochs, batch_size=batch_size)
history_lstm = model_lstm.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=epochs, batch_size=batch_size)
history_bi_lstm = model_bi_lstm.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=epochs, batch_size=batch_size)
history_deep_lstm = model_deep_lstm.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=epochs, batch_size=batch_size)
```

```python
import matplotlib.pyplot as plt

def plot_history(histories, title):
    plt.figure(figsize=(12, 8))
    
    for name, history in histories.items():
        plt.plot(history.history['val_accuracy'], label=f'{name} val_acc')
        plt.plot(history.history['accuracy'], label=f'{name} acc')

    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

histories = {
    'RNN': history_rnn,
    'LSTM': history_lstm,
    'Bidirectional LSTM': history_bi_lstm,
    'Deep LSTM': history_deep_lstm
}

plot_history(histories, 'Model Accuracy Comparison')
```

На графіку видно, що найбільш високу точність як на тренувальних, так і на валідаційних даних показали моделі з LSTM, зокрема двостороння та глибока LSTM. Вони демонструють стабільно високі результати, перевищуючи 90% точності вже після перших кількох епох.

Натомість звичайна RNN демонструє значно нижчі результати як на тренувальних, так і на валідаційних даних. Вона показує нестабільність і невелику точність (приблизно 70%).

LSTM моделі перевершують RNN у цьому завданні, оскільки краще обробляють довгі послідовності даних.
Двостороння та глибока LSTM ще більше покращують результати порівняно зі звичайною LSTM, що свідчить про те, що збільшення складності моделей виправдовується на великих послідовних датасетах.
